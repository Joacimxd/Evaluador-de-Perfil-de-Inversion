```python
import kmeans
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("sharpe_cagr.csv", index_col=0)
df.iloc[:, 1:3] = kmeans.normalize_per_col(df.iloc[:, 1:3])
df.values.shape[1]
```

```python
k_means = kmeans.KMeans(10, linear=True)
k_means.train(df)
plt.scatter(df.values[:,1], df.values[:,2])
plt.scatter(k_means.means[:,0], k_means.means[:, 1])
```
